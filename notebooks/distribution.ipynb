```python
from collections import Counter
import seaborn as sns
import pandas as pd
```

```python
def get_basic_info(dataset):
    counter = Counter()
    for idx, x in dataset.iterrows():
        labels = x.labels.split(',')
        counter.update(labels)
    
    print('Number of training examples: ', dataset.shape[0])
    print('Number of unique classes: ', len(counter))
```

```python
def get_labels_graph(counter):
    keys = list(counter.keys())
    values = list(counter.values())
    data_dict = {'label': keys, 'number': values}
    data = pd.DataFrame.from_dict(data_dict)
    sns.catplot(x='number', y="label", kind="bar", data=data, height=20);
```

```python
def get_occurences_graph(counter):
    keys = list(counter.keys())
    values = list(counter.values())
    data_dict = {'num_labels': keys, 'num_examples': values}
    data = pd.DataFrame.from_dict(data_dict)
    sns.catplot(x='num_labels', y='num_examples', kind='bar', data=data)
```

```python
def get_labels_counter(dataset):
    counter = Counter()
    for idx, x in dataset.iterrows():
        labels = x.labels.split(',')
        counter.update(labels)
    
    return counter
```
